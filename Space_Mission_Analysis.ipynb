```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df=pd.read_csv('./DataSets/Space_Corrected.csv')
df
```

```python
df.isnull().sum()
```

```python
df['Price']= df['Price'].fillna(0)
df
```

```python
df=df.rename(columns={'Detail':'Rocket'})
df
```

```python
df=df.rename(columns={' Rocket':'Price'})
df
```

```python
df=df.rename(columns={'Datum':'Datetime'})
df
```

```python
df['Year']=df.Datetime.str[11:16]
df
```

```python
df.columns
```

```python
df['month']=df.Datetime.str[3:7]
df
```

```python
df['Year']=df['Year'].replace('[.,]','',regex=True).astype('int64')
df.Year=df.Year.replace('','0',regex=True)
df.Year=df.Year.astype('int64')
df
```

```python
c = df['Company Name'].value_counts().nlargest(10)

plt.figure(figsize=(10,5))
sb.barplot(x=c.index, y=c.values, color='#FF1C00')  
plt.xlabel('Company Name', fontsize=12)
plt.ylabel('Number of Launches', fontsize=12)
plt.title('Top Companies by Launch Count', fontsize=14, fontweight='bold')
plt.xticks(rotation=90)
plt.show()
```

```python
Succesful_Missions=df[df['Status Mission'] == 'Success'].groupby('Company Name').size().reset_index(name='count').sort_values(by='count',ascending=False)
Succesful_Missions
```

```python
df.Price=df.Price.replace(',','',regex=True)
df
```

```python
df.Price=df.Price.astype('float64')
```

```python
g=df['Location'].value_counts().nlargest(10)

plt.figure(figsize=(10,5))
sb.barplot(x=g.index, y=g.values, color='#FF0000') 
plt.xlabel('Launch Location', fontsize=12)
plt.ylabel('Number of Launches', fontsize=12)
plt.title('Top Launch Locations', fontsize=14, fontweight='bold')
plt.xticks(rotation=90)
plt.show()
```

```python
m= df[df['Status Mission'] == 'Success'].groupby('Location').size().nlargest(10)

plt.figure(figsize=(10,5))
sb.barplot(x=m.index, y=m.values, color='#FFD300') 
plt.xlabel('Launch Location', fontsize=12)
plt.ylabel('Number of Launches', fontsize=12)
plt.title('Top Space Missions by Launch Count', fontsize=14, fontweight='bold')
plt.xticks(rotation=90)
plt.show()
```

```python
s=df.Price.value_counts().nlargest(10)

plt.figure(figsize=(10,5))
sb.barplot(x=s.index, y=s.values, color='#FF7043')  
plt.xlabel(' Rocket Name', fontsize=12)
plt.ylabel('Number of Launches', fontsize=12)
plt.title('Most Frequently Used Rockets', fontsize=14, fontweight='bold')
plt.xticks(rotation=90)
plt.show()
```
